```
#Następny krok to wczytanie odpowiednich bibliotek

from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
```

```
#Następnie musimy wczytać nasze dane
#Delimiter jest to separator, który oddziela od siebie dane, w tym przypadku jest to ","
dataset = loadtxt('bank-full-opracowane.csv', delimiter=',')
```

```
#Podzielimy nasze dane na zbiór X oraz Y
#X to będą wartości, które dostarczamy do naszej sieci - będzie to 16 cech
#Y to zbiór wyników - czyli informacja, czy klient skorzysta z oferty banku
X = dataset[:,0:16]
Y = dataset[:,16]
```

```
#Teraz przechodzimy do tworzenia modelu w Keras
#Pierwsza linijka oznacza inicjację tworzenia naszego modelu
#Druga linijka odpowiada za dodanie warstw, pierwsza liczba to warstwa ukryta, ma 12 neuronów, warstwa wejściowa ma 16 neuronów, ponieważ tyle cech mają nasze dane, funkcja aktywacji to 'relu'
#Trzecia linijka odpowiada za dodanie drugiej warstwy ukrytej, składa się z 16 neuronów, aktywacja to 'relu'
#Ostatnia linijka odpowiada za warstwę wyjściową, składa się tylko z 1 neuronu, ponieważ ona da nam odpowiedź czy klient skorzysta z banku czy nie


model = Sequential()
model.add(Dense(12, input_dim=16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```
#Przechodzimy do kompilacji modelu, korzystamy z najczęściej stosowanych ustawień

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
#Kolej na trening naszej sieci
#Przekazujemy do modelu nasze cechy (było ich 16) i informujemy nasz model, jaki wynik jest dla tych cech wyjściowych
#Sieć korzysta z danych i etykiet, dzięki temu się uczy
# Epchos to liczba iteracji - powtórzeń
#Przypisuje całość do zmiennej history, ponieważ przyda się do sprawdzenia skuteczności sieci

history = model.fit(X, Y, epochs=150, batch_size=10, verbose=2)
```

```
#Po wytrenowaniu naszego modelu, wpisujemy kod odpowiadający za przewidywania od naszej sieci

predictions = model.predict_classes(X)
```

```
#Teraz wypiszemy wyniki dla 10 obserwacji
#Korzystam z pętli for z udostępnionego wcześniej przykładu

for i in range(10):
	print('cechy wejściowe obserwacji (input): %s => przewidziany wynik (output): %d (rzeczywista wartość: %d)' % (X[i].tolist(), predictions[i], Y[i]))
```

```
#Teraz pozyskam wskaźnik skuteczności naszej sieci
#Przy pomocy .history

history.history['accuracy']
```

```
#Teraz chcę otrzymać info o skuteczności całego naszego modelu
#Korzystam z metody .evaluate

model.evaluate(X,Y)
```
